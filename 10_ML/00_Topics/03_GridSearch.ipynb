# GridSearchCV

searchs for the best model hyper param

```python
from sklearn.datasets import load_digits, load_iris

from sklearn.linear_model import LogisticRegression

import pandas as pd 

from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedStratifiedKFold

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

# Load the Data

```python
iris = load_iris()
```

```python
df = pd.read_csv("../datasets/covid.csv")

X  = df[["age"]]
y  = df[["covid"]]

X_train, X_test, y_train, y_test = train_test_split(X, y , train_size= 0.8)
```

# Manual Searching

```python
# model = LogisticRegression(solver= "lbfgs")
model = LogisticRegression(solver= "liblinear")

model.fit(X_train, y_train)
```

# Searching using GridSearchCV

```python
# 0. Create the instance of the model
model = LogisticRegression()
```

```python
# 1. Define the parameter Grid

param_grid = {
    "solver": ["lbfgs" , "liblinear"],
    "C" : [100, 10, 1.0, 0.1]
}

```

```python
# 2. Define the K-Folds for Cross Validation
cv = RepeatedStratifiedKFold(n_splits= 10, n_repeats= 3)
```

```python
# 3. Define the Grid Search
grid_search = GridSearchCV(estimator= model, param_grid= param_grid, cv = cv, scoring = "accuracy")
```

# Train the model

```python
grid_result = grid_search.fit(iris.data, iris.target)
```

# Show the best model score and best param combi

```python
print(grid_result.best_score_)
print(grid_result.best_params_)
```

```python
results_df = pd.DataFrame(grid_search.cv_results_)
results_df
```

# GridSearchCV to test also several models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = RepeatedStratifiedKFold(n_splits = 10 , n_repeats= 3)
```

```python
model_param_grid = {
    "logisitc_regression": {
        "model": LogisticRegression(),
        "param_grid": {
                        "solver": ["lbfgs" , "liblinear"],
                        "C" : [100, 10, 1.0, 0.1]
                     }
    },
    "randomforest": {
        "model": RandomForestClassifier(),
        "param_grid": {
            "n_estimators": [10, 50 , 100 ,500]
        }
    }
}

from pprint import pprint
for model_name, model_param in model_param_grid.items():

    pprint(model_name)
    pprint(model_param_grid)
```

```python
# collect the scores cross all trainings
scores = []

for model_name, model_param in model_param_grid.items():
    # grid_search = GridSearchCV(estimator=model_param["model"] , param_grid= model_param["param_grid"], cv = cv)
    grid_search = RandomizedSearchCV(estimator=model_param["model"] , param_distributions= model_param["param_grid"], cv = cv, n_iter= 3)

    grid_search.fit(iris.data, iris.target)

    scores.append({
        "model": model_name,
        "best_score" : grid_search.best_score_,
        "best_param" : grid_search.best_params_,
    })
```

```python
df_multi_grid = pd.DataFrame(scores, columns= ["model", "best_score", "best_param"])

df_multi_grid
```
