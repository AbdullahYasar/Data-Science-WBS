```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
```

```python
df = pd.read_csv("./datasets/dataset.csv")
df.head()
```

```python
plt.scatter(df.area, df.price)
plt.xlabel("Area")
plt.ylabel("Price")

plt.grid()
```

```python
# Input Data
input_df = df.drop("price", axis = "columns")

# Output Data
price = df.price
```

```python
input_df.head()

price
```

```python
model = linear_model.LinearRegression()

model.fit(input_df, price)
```

```python
print(model.coef_)  # slope (m)
print(model.intercept_) # intercept (b)
```

```python
area = 70

area_price = area * model.coef_ + model.intercept_

print(area_price)
```

```python
model.predict([[70], [90], [100]])
model.predict([[80]])
```

```python
prediction_df = pd.read_csv("./datasets/dataset_new.csv")

prediction_df.head()
```

```python
# Calculate the prices

predicted_prices = model.predict(prediction_df)

predicted_prices

prediction_df["predicted_pricest"] = predicted_prices

```
